```python
import pandas as pd
import re
# Leemos el CSV con los códigos por provincia eliminando las 5 primeras columnas que ya
# hemos visto que no son relevantes.
bizkaia = pd.read_csv('Bizkaia.csv', index_col=False,encoding='iso-8859-1',sep=';',skiprows=5)
bizkaia.head(10)
```

```python
# Renombramos las colimnas para ponerles nomnbres normales.
bizkaia.rename(columns={'Unnamed: 0':'identificador','2016':'renta'},inplace=True)
bizkaia
```

```python
# Nos deshacemos de las últimas filas que tampoco son relevantes.
bizkaia.drop(bizkaia.tail(3).index,inplace=True)
bizkaia
```

```python
# Función para obtener el NIE de la BBDD (base de datos).
def get_NIE (bbdd):
    NIE = []
    for i in range(len(bbdd.identificador)):
        NIE_value = bbdd.identificador[i][0:9]
        NIE.append(NIE_value)
    return NIE
NIE = get_NIE(bizkaia)


```
